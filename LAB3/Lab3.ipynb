```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#2 wykresy + przestrzeń między nimi
fig, ax_lst = plt.subplots(2)  # a figure with a 2x2 grid of Axes
fig.subplots_adjust(left=None, bottom=None, right=None, top = 2)

#utworz 2 ciągi liczb gdzie jedna będzie miała wartości od 0 do 2 i pomiedzy nimi będzie 98 licz (iteruj od 0 do 2)
x = np.linspace(0, 1, 100)
y = np.linspace(0, 5, 100)
#narysowanie wyniku
ax_lst[0].plot(x, y , label='Wynik')
#ustawienie nazw poziom, pion osi i nazwa wykresu
ax_lst[0].set(xlabel = 'ox') 
ax_lst[0].set(ylabel = 'oy')  
ax_lst[0].set_title("flin ") 
ax_lst[0].legend(loc="upper right")

#narysuj wykres dla funkcji -x * 5
b = 4
x = range(0, 10)  # lista argumentów x
y = []  # lista wartości
for i in x:
    y.append((1 * i) * 4)
    
ax_lst[1].plot(x, y , label='Wynik')
#ustawienie nazw poziom, pion osi i nazwa wykresu
ax_lst[1].set_title('Wykres f(x) = x + 4')
ax_lst[1].grid(True)
ax_lst[1].legend(loc="upper right")
```

```python
#2 wykresy + przestrzeń między nimi
fig, ax_lst = plt.subplots(2)  # a figure with a 2x2 grid of Axes
fig.subplots_adjust(left=None, bottom=None, right=None, top = 2)

#2 ciągi liczb jeden z  100 elementami pomiędzy  -10 a 10. 
x = np.linspace(-10,10,100)
y = x**2
#narysuj wykres funkcj y=x^2 
ax_lst[0].plot(x,y, 'r',label='Wynik')
#ustawienie nazw poziom, pion osi i nazwa wykresu
ax_lst[0].set(xlabel = 'ox') #ustaw x label czyli to co na dole
ax_lst[0].set(ylabel = 'oy')  #ustaw x label czyli to co po prawej
ax_lst[0].set_title("Fkwadratowa") #ustaw tytuł osi
ax_lst[0].legend(loc="upper right")
#Funkcja kwadratowa 2 
x = np.linspace(0, 10, 100)
ax_lst[1].plot(x, x**2,label='Wynik')
#ustawienie nazw poziom, pion osi i nazwa wykresu
ax_lst[1].set(xlabel = 'ox') 
ax_lst[1].set(ylabel = 'oy') 
ax_lst[1].set_title("FKwadratowa") 
ax_lst[1].legend(loc="upper right")
```

```python
#2 wykresy + przestrzeń między nimi
fig, ax_lst = plt.subplots(2)  # a figure with a 2x2 grid of Axes
fig.subplots_adjust(left=None, bottom=None, right=None, top = 2)

#utworz 3 ciągi liczb jeden z  100 elementami pomiędzy  -10 a 10. Drugi który bedzie posiadał wartości podniesione do potęgi 3
##tzn y = x^3
x = np.linspace(-10,10,100)
y = x**3
#na podstawie funkcj y=x^3 narysuj wykres 
ax_lst[0].plot(x,y, 'r',label='Wynik')
#ustawienie nazw poziom, pion osi i nazwa wykresu
ax_lst[0].set(xlabel = 'ox') 
ax_lst[0].set(ylabel = 'oy') 
ax_lst[0].set_title("F trzeciego rzędu ") 
ax_lst[0].legend(loc="upper right")
#Funkcja Trzeciego rzędu 2 
x = np.linspace(0, 10, 100)
ax_lst[1].plot(x, x**3, label='Wynik')
#ustawienie nazw poziom, pion osi i nazwa wykresu
ax_lst[1].set(xlabel = 'ox') 
ax_lst[1].set(ylabel = 'oy')  
ax_lst[1].set_title("F trzeciego rzędu ") 
ax_lst[1].legend(loc="upper right")
```

```python
#Naryuj 2 subloty i ustaw przestrzeń pomiędzy nimi 
fig, ax_lst = plt.subplots(2)  # a figure with a 2x2 grid of Axes
fig.subplots_adjust(left=None, bottom=None, right=None, top = 2)

#utworz 2 ciągi jeden z liczbami rozmieszczonymi liniowo z użyciem pi drugi y = sin(x)
x = np.linspace(-np.pi,np.pi,100)
y = np.sin(x)

#na podstawie funkcj y=sin(x) narysuj wykres 
ax_lst[0].plot(x,y, 'r',label='Wynik')
ax_lst[0].set(xlabel = 'ox') #ustaw x label czyli to co na dole
ax_lst[0].set(ylabel = 'oy')  #ustaw x label czyli to co po prawej
ax_lst[0].set_title("F trygonometryczna") #ustaw tytuł osi
ax_lst[0].legend(loc="upper right")

#Utworz liczby od 0 do 100 inkrementowane co 0.4 i oblicz sinus każdego z nich
x = np.arange(0, 100, .4)
y = np.sin(x)
ax_lst[1].plot(x, y,label='Wynik')
ax_lst[1].set(xlabel = 'ox') #ustaw x label czyli to co na dole
ax_lst[1].set(ylabel = 'oy')  #ustaw x label czyli to co po prawej
ax_lst[1].set_title("F trygonometryczna") #ustaw tytuł osi
ax_lst[1].legend(loc="upper right")
```

```python
# z wykorzystaniem pandy
cars = pd.read_csv('samochody1tys.csv') #pobierz dane
df = cars.marka.value_counts().to_frame(); #policz ilości marek a następnie zapisz to do obiektu frame
df.plot.bar(y='marka') # narysuj na podstawie ilości marek 
plt.ylabel('Ilość')
plt.xlabel('Marki')
plt.title('Ilość marek')
```

```python
#Wykres kołowy pprzedstawiający ilość aut wyprodukowanych w danych latach oraz ich procent co do całości
#pobranie wartości roku producji i zapis do listy
labels  = cars['rok_produkcji'].value_counts().keys().tolist()
#pobierz ilość wystąpień roku producji i zapisz do listy
sizes  = cars['rok_produkcji'].value_counts().tolist()

#narysowanie wykresu ilości wystąpień i obrócenie go o 90 stopni
patches, texts = plt.pie(sizes, startangle=90)
plt.legend(patches, labels) #utworzenie legendy na podstawie lat
plt.axis('equal') # zmiana położenia wykresu
plt.ylabel('Ilość')
plt.xlabel('Wykres')
plt.title('Ilość roków produkcji')
plt.show()
```

```python
#histogram na podstawie przebiegów + 50 krawędzi dla lepszego odczytania
przebieg  = cars['przebieg'].value_counts().keys().tolist()
plt.hist(przebieg,50,label='Ilości wystąpień przebiegów')
plt.ylabel('Przebieg')
plt.xlabel('Ilość')
plt.title('Ilość przebiegów ')
plt.legend(loc="upper right")
plt.show()
```
